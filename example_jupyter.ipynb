```python
from pathlib import Path

import solara
from hdxms_datasets import DataBase

from instagibbs.preprocess import load_hdx_dataset
from instagibbs.web.app import DataManager, ViewCard
```

```python
root = Path().resolve()
```

```python
database_dir = root / "tests" / "test_data"
vault = DataBase(database_dir=database_dir)

DATASET = "HDX_0A55672B"
ds = vault.load_dataset(DATASET)
```

```python
states = load_hdx_dataset(ds)
# row of viewcards; shared datamanager

dm = DataManager(hdx_states=states)
solara.Row(
    (ViewCard(dm, view_type='peptides_chart', height='500px'),
    ViewCard(dm, view_type='residues_chart', height='500px'),
    ViewCard(dm, view_type='structure', height='500px'),
    )
)
```

```python
# row of viewcards; individual DataManagers
solara.Row(
    (ViewCard(DataManager(hdx_states=states), height='500px'),
    ViewCard(DataManager(hdx_states=states), height='500px'))
)
```
